```python
# Spell Checking using Minimum Edit Distance

def edit_distance(s1, s2):
    if s1 == s2:
        return 0
    elif abs(len(s1) - len(s2)) == 1:
        return 1
    else:
        count = 0
        for i in range(min(len(s1), len(s2))):
            if s1[i] != s2[i]:
                count += 1
        count += abs(len(s1) - len(s2))
        return count

# Correct words
words = ["apple", "banana", "orange", "python", "computer"]

# Wrong word
word = input("Enter a word: ")

min_distance = 100
correct_word = ""

for w in words:
    d = edit_distance(word, w)
    if d < min_distance:
        min_distance = d
        correct_word = w

print("Suggested Word :", correct_word)
print("Edit Distance  :", min_distance)
```
