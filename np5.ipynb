```python
# Simple Co-reference Resolution

text = "Rahul is a student. He studies Python. Rahul likes programming."

words = text.split()

name = ""
result = []

for word in words:
    clean = word.strip(".,!?")

    # Save the person's name
    if clean == "Rahul":
        name = clean
        result.append(clean)
    # Replace pronouns with the name
    elif clean in ["He", "he", "His", "his"]:
        result.append(name)
    else:
        result.append(clean)

print("Original Text:")
print(text)

print("\nAfter Co-reference Resolution:")
print(" ".join(result))
```
