```python
import re


text = "Hello Friends! Welcome to NLP Practical. I am Learning Text Preprocessing in Python, 2026."

print("Original Text:")
print(text)



# Convert text to lowercase
text = text.lower()
print("\nLowercase Text:")
print(text)

# Remove punctuation
text = re.sub(r'[^\w\s]', '', text)
print("\nAfter Removing Punctuation:")
print(text)



# Split text into words
tokens = text.split()

print("\nTokens:")
print(tokens)

# Count total words
print("\nTotal Words:", len(tokens))

# Display each token
print("\nWords in Text:")
for word in tokens:
    print(word)

# Join words again
new_text = " ".join(tokens)

print("\nFinal Preprocessed Text:")
print(new_text)
```
