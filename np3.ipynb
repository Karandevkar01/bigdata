```python
# Viterbi Algorithm for POS Tagging

tags = ["Noun", "Verb"]
words = ["Dogs", "bark"]

start = {"Noun": 0.5, "Verb": 0.5}

transition = {
    "Noun": {"Noun": 0.4, "Verb": 0.6},
    "Verb": {"Noun": 0.5, "Verb": 0.5}
}

emission = {
    "Noun": {"Dogs": 0.8, "bark": 0.2},
    "Verb": {"Dogs": 0.1, "bark": 0.9}
}

result = []

# First word
if start["Noun"] * emission["Noun"][words[0]] > start["Verb"] * emission["Verb"][words[0]]:
    result.append("Noun")
    current = "Noun"
else:
    result.append("Verb")
    current = "Verb"

# Remaining words
for i in range(1, len(words)):
    n = transition[current]["Noun"] * emission["Noun"][words[i]]
    v = transition[current]["Verb"] * emission["Verb"][words[i]]

    if n > v:
        result.append("Noun")
        current = "Noun"
    else:
        result.append("Verb")
        current = "Verb"

print("Words :", words)
print("POS Tags :", result)
```
