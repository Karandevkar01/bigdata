```python
# A* algo
from queue import PriorityQueue

# Maze
maze = [
    ["S", ".", ".", "#", "."],
    [".", "#", ".", "#", "."],
    [".", "#", ".", ".", "."],
    [".", ".", "#", "#", "."],
    ["#", ".", ".", "G", "."]
]

rows = len(maze)
cols = len(maze[0])

start = (0, 0)
goal = (4, 3)

# Up, Down, Left, Right
move = [(-1,0), (1,0), (0,-1), (0,1)]

# Find distance
def distance(a, b):
    return abs(a[0]-b[0]) + abs(a[1]-b[1])

pq = PriorityQueue()
pq.put((0, start))

came = {}
cost = {}
cost[start] = 0

while not pq.empty():
    now = pq.get()[1]

    if now == goal:
        break

    for m in move:
        new = (now[0]+m[0], now[1]+m[1])

        if 0 <= new[0] < rows and 0 <= new[1] < cols:
            if maze[new[0]][new[1]] != "#":

                new_cost = cost[now] + 1

                if new not in cost or new_cost < cost[new]:
                    cost[new] = new_cost
                    total = new_cost + distance(new, goal)
                    pq.put((total, new))
                    came[new] = now

# Print path
path = []
p = goal

while p != start:
    path.append(p)
    p = came[p]

path.append(start)
path.reverse()

print("Shortest Path:")
print(path)
```

```python
from queue import PriorityQueue

# Maze
maze = [
    ["S", ".", ".", "#", "."],
    [".", "#", ".", "#", "."],
    [".", "#", ".", ".", "."],
    [".", ".", "#", "#", "."],
    ["#", ".", ".", "G", "."]
]

rows = len(maze)
cols = len(maze[0])

start = (0, 0)
goal = (4, 3)

move = [(-1,0), (1,0), (0,-1), (0,1)]

# Find distance
def distance(a, b):
    return abs(a[0]-b[0]) + abs(a[1]-b[1])

pq = PriorityQueue()
pq.put((distance(start, goal), start))

visited = set()
visited.add(start)

came = {}

while not pq.empty():
    now = pq.get()[1]

    if now == goal:
        break

    for m in move:
        new = (now[0]+m[0], now[1]+m[1])

        if 0 <= new[0] < rows and 0 <= new[1] < cols:
            if maze[new[0]][new[1]] != "#" and new not in visited:
                visited.add(new)
                pq.put((distance(new, goal), new))
                came[new] = now

# Print path
path = []
p = goal

while p != start:
    path.append(p)
    p = came[p]

path.append(start)
path.reverse()

print("Path Found:")
print(path)
```
