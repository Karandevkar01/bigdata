```python
# Sudoku Solver using Backtracking

board = [
    [5,3,0,0,7,0,0,0,0],
    [6,0,0,1,9,5,0,0,0],
    [0,9,8,0,0,0,0,6,0],
    [8,0,0,0,6,0,0,0,3],
    [4,0,0,8,0,3,0,0,1],
    [7,0,0,0,2,0,0,0,6],
    [0,6,0,0,0,0,2,8,0],
    [0,0,0,4,1,9,0,0,5],
    [0,0,0,0,8,0,0,7,9]
]

# Find empty place
def find_empty():
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                return (i, j)
    return None

# Check number is safe
def is_safe(row, col, num):

    # Check row
    for i in range(9):
        if board[row][i] == num:
            return False

    # Check column
    for i in range(9):
        if board[i][col] == num:
            return False

    # Check 3x3 box
    start_row = row - row % 3
    start_col = col - col % 3

    for i in range(3):
        for j in range(3):
            if board[start_row+i][start_col+j] == num:
                return False

    return True

# Backtracking
def solve():

    empty = find_empty()

    if empty is None:
        return True

    row, col = empty

    for num in range(1, 10):

        if is_safe(row, col, num):
            board[row][col] = num

            if solve():
                return True

            board[row][col] = 0

    return False

solve()

print("Solved Sudoku:")

for row in board:
    print(row)
```

```python
# Map Coloring using Backtracking

graph = {
    "A": ["B", "C"],
    "B": ["A", "C", "D"],
    "C": ["A", "B", "D"],
    "D": ["B", "C"]
}

colors = ["Red", "Green", "Blue"]

result = {}

# Check color
def is_safe(place, color):

    for neighbour in graph[place]:
        if neighbour in result and result[neighbour] == color:
            return False

    return True

# Backtracking
def solve(place_list, index):

    if index == len(place_list):
        return True

    place = place_list[index]

    for color in colors:

        if is_safe(place, color):
            result[place] = color

            if solve(place_list, index + 1):
                return True

            del result[place]

    return False

places = list(graph.keys())

solve(places, 0)

print("Map Coloring:")

for place in result:
    print(place, "=", result[place])
```
