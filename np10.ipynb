```python
import pandas as pd
df = pd.read_csv("Customer_Sentiment.csv")
```

```python
df
df =df.drop('sentiment',axis=1)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['review']=df['review_text'].str.lower()
```

```python
positive =["good","great","excellent","love","Happy"]
nigative=["bad","poor","worst","hate","problame","slow"]
```

```python
def analysis(review):
    review=str(review).lower()
    pos= sum(word in review.split()for word in positive)
    nig=sum(word in review.split()for word in nigative)

    if pos> nig:
        return "positive"
    elif nig> pos:
        return "positive"
    else:
        return "natural"
        
```

```python
df["sentiment"]=df["review"].apply(analysis)
print(df[["review","sentiment"]])
```
