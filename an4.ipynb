```python
# Medical Diagnosis using First-Order Logic

# Facts
facts = {
    "Fever": ["John"],
    "Cough": ["John"],
    "Flu": [],
    "NeedsRest": [],
    "SevereCough": [],
    "ShortnessOfBreath": [],
    "Pneumonia": []
}

# Rule 1: Fever(x) AND Cough(x) -> Flu(x)
for person in facts["Fever"]:
    if person in facts["Cough"]:
        facts["Flu"].append(person)

# Rule 2: Flu(x) -> NeedsRest(x)
for person in facts["Flu"]:
    facts["NeedsRest"].append(person)

# Rule 3: SevereCough(x) AND ShortnessOfBreath(x) -> Pneumonia(x)
for person in facts["SevereCough"]:
    if person in facts["ShortnessOfBreath"]:
        facts["Pneumonia"].append(person)

# Print Results
print("Inference Results")
print("-----------------")

if facts["Flu"]:
    print("Flu:", facts["Flu"])

if facts["NeedsRest"]:
    print("Needs Rest:", facts["NeedsRest"])

if facts["Pneumonia"]:
    print("Pneumonia:", facts["Pneumonia"])
else:
    print("No Pneumonia")
```
