```python
import numpy as np
from sklearn.linear_model import LinearRegression, SGDRegressor

# House Data
# Size, Rooms
X = np.array([
    [1000, 2],
    [1200, 3],
    [1500, 3],
    [1800, 4],
    [2000, 4]
])

# House Prices
y = np.array([20, 25, 30, 35, 40])


# Batch Gradient Descent

bgd = LinearRegression()
bgd.fit(X, y)

print("Batch Gradient Descent")
print("Intercept:", bgd.intercept_)
print("Weights:", bgd.coef_)


# Stochastic Gradient Descent

sgd = SGDRegressor(max_iter=1000, learning_rate="constant", eta0=0.000001)
sgd.fit(X, y)

print("\nStochastic Gradient Descent")
print("Intercept:", sgd.intercept_)
print("Weights:", sgd.coef_)


# Mini-Batch Gradient Descent

mbgd = SGDRegressor(max_iter=1000, learning_rate="constant", eta0=0.000001)

batch_size = 2

for epoch in range(100):
    for i in range(0, len(X), batch_size):
        X_batch = X[i:i+batch_size]
        y_batch = y[i:i+batch_size]
        mbgd.partial_fit(X_batch, y_batch)

print("\nMini-Batch Gradient Descent")
print("Intercept:", mbgd.intercept_)
print("Weights:", mbgd.coef_)
```
