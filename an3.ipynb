```python
# Medical Diagnosis System using First-Order Logic

# Get symptoms from user
symptoms = []

print("Enter Symptoms")
print("Options: fever, cough, headache, bodypain")
print("Type 'done' to stop")

while True:
    s = input("Symptom: ")

    if s == "done":
        break

    symptoms.append(s)

# FOL Rules
if "fever" in symptoms and "cough" in symptoms:
    print("Disease: Flu")

elif "fever" in symptoms and "bodypain" in symptoms:
    print("Disease: Dengue")

elif "cough" in symptoms and "headache" in symptoms:
    print("Disease: Covid-19")

elif "cough" in symptoms:
    print("Disease: Common Cold")

else:
    print("Disease: Disease Not Found")
```
